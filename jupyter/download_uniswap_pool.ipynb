```python
import hyperline
import pandas as pd
```

```python
# Get all pool stats for Uniswap V3 contract https://etherscan.io/address/0x8ad599c3a0ff1de082011efddc58f1908eb6e6d8 
sql = """ SELECT * FROM ethereum.uniswapv3.ez_pool_stats WHERE pool_address = '0x8ad599c3a0ff1de082011efddc58f1908eb6e6d8' """
df = hyperline.sql.execute_flipside_sql(sql)
```

```python
with pd.option_context('display.max_colwidth', None):
  display(df)
```

```python
location = hyperline.storage.path()
# Save as parquet
df.to_parquet(location + 'uniswap_pool_stats.parquet')
```
